```python
import pandas as pd 
from wordcloud import WordCloud

def minimal_wordcloud(df, column):
    
    text = str(df[column].values)
    wordcloud = WordCloud().generate(text)
    image = wordcloud.to_image()
    return image
```

```python
recipe_df = pd.read_csv('RecipeNLG_dataset.csv')
```
